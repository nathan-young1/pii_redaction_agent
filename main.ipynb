### Implementing and testing the redaction workflow on a sample PDF

```python
import textwrap
from typing import Optional
from pathlib import Path

from agents import (
    Agent,
    GuardrailFunctionOutput,
    Runner,
    output_guardrail,
    OutputGuardrailTripwireTriggered,
    AsyncOpenAI, 
    OpenAIChatCompletionsModel, 
    set_tracing_disabled,
    ModelSettings,
)
from pydantic import BaseModel, Field

set_tracing_disabled(disabled=True) # Not needed for this example.

client = AsyncOpenAI(
    api_key="sk-no-key-required", 
    base_url="http://127.0.0.1:8000/v1",
)

MODEL = OpenAIChatCompletionsModel(model="Qwen3.6-35B-A3B-UD-IQ1_M.gguf", openai_client=client)
MODEL_SETTINGS = ModelSettings(
        temperature=0.7, 
        max_tokens=8192, 
        extra_body={
        "top_k": 20,
        "chat_template_kwargs": {"enable_thinking": False}, # fairly simply task, so we can disable the "thinking".
        }
    )
```

```python
class RedactionOutput(BaseModel):
    redacted_markdown: str = Field(description="The markdown output after PII redaction.")


class ReviewOutput(BaseModel):
    approved: bool
    rejection_reason: Optional[str] = Field(description="If not approved, a brief explanation of why.")
    remaining_pii_if_rejected: Optional[list[str]] = Field(description="List of any PII snippets still present, only if document was rejected.")



# Review agent (used as the redaction output guardrail).
guardrail_review_agent = Agent(
    name="ReviewAgent",
    model=MODEL,
    model_settings=MODEL_SETTINGS, 
    output_type=ReviewOutput,
    instructions=textwrap.dedent(f"""
        You are a strict PII audit agent. Check whether the redacted document
        passed to you still contains any remaining personally identifiable information.

        Note: 
          - Also reject documents with Quasi-identifiers (combinations like job title + employer + city
            that together uniquely identify a real person)

        Return your result in the structured output format provided.
    """).strip(),
)


@output_guardrail
async def review_guardrail(ctx, agent, output: RedactionOutput) -> GuardrailFunctionOutput:
    """
    Output guardrail — ReviewAgent runs here automatically after RedactionAgent produces output, before it is returned
    """
    result = await Runner.run(guardrail_review_agent, output.redacted_markdown, context=ctx.context)
    review = result.final_output_as(ReviewOutput)

    print(f"  {'✅ Review PASSED' if review.approved else '❌ Review FAILED - ' + review.rejection_reason}")

    # Trigger the guardrail tripwire if the review agent did not approve the redaction.
    return GuardrailFunctionOutput(
        output_info=review,
        tripwire_triggered=not review.approved,
    )



# Redaction agent with guardrail attached
redaction_agent = Agent(
    name="RedactionAgent",
    model=MODEL,
    model_settings=MODEL_SETTINGS,
    output_type=RedactionOutput,
    output_guardrails=[review_guardrail],
    instructions=textwrap.dedent(f"""
        You are a PII (Personal Identifiable Information) redaction agent. 
        Replace all PII (Personal Identifiable Information) with placeholders like [NAME], [EMAIL], [PHONE] etc.

        Rules:
          - Do not redact standalone company names or city names unless it is paired
            with a personal identifier.
          - Also redact Quasi-identifiers (combinations like job title + employer + city
            that together uniquely identify a real person).
          - Preserve all punctuation, spacing, table and document structure.

        Return your result in the structured output format provided.           
    """).strip(),
)


def build_redaction_prompt(original_text: str, attempt: int, feedback: Optional[str] = None) -> str:
    if attempt == 0: # First redaction attempt.
        return f"Redact all PII from the following text:\n\n{original_text}"
    else:
        return textwrap.dedent(f"""
            Your previous redaction was rejected. Reviewer feedback:
            {feedback}

            Re-redact the ORIGINAL content below more carefully:
            {original_text}
        """).strip()


def write_output_to_file(redaction_text: RedactionOutput, output_file: Path):
    output_file.write_text(redaction_text, encoding="utf-8")
    print(f"\n✅ Written to: {output_file}")


async def perform_redaction(original_content: str, output_file: Path, max_attempts: int = 3):

    feedback = None

    for attempt in range(max_attempts):
        print(f"[{attempt}/{max_attempts}] RedactionAgent running...")

        try:
            result = await Runner.run(
                redaction_agent,
                build_redaction_prompt(original_content, attempt, feedback)
            )

            # Reaching here means the guardrail passed
            output = result.final_output_as(RedactionOutput)
            write_output_to_file(output.redacted_markdown, output_file)
            return

        except OutputGuardrailTripwireTriggered as e:
            # Guardrail tripped — extract reviewer feedback for next attempt
            review: ReviewOutput = e.guardrail_result.output.output_info
            
            feedback = review.model_dump_json()

            if attempt < (max_attempts - 1):
                print(f"  Retrying with feedback...\n")
            else:
                print(f"\n❌ Aborted after {max_attempts} attempts. Manual review required.")
```

```python
# convert pdf file to md
import pymupdf4llm

def convert_file_to_md(file_path: Path) -> str:
    md = pymupdf4llm.to_markdown(file_path)
    return md
    
```

```python
# Texting the redaction on a sample pdf.
INPUT_FILE = Path("medical_pii_sample.pdf")
OUTPUT_FILE = Path("redacted_medical_sample.md")


content_as_markdown = convert_file_to_md(INPUT_FILE)

# Perform redaction.
await perform_redaction(
        original_content=content_as_markdown,
        output_file=OUTPUT_FILE,
        max_attempts=3
    )
```
